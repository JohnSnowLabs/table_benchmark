```
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
import mmcv
# Load model
config_file = '/content/mmdetection/CascadeTabNet/Config/cascade_mask_rcnn_hrnetv2p_w32_20e.py'
# checkpoint_file = '/content/mmdetection/epoch_36.pth' # TSR
checkpoint_file = '/content/mmdetection/epoch_14.pth' # TD
model = init_detector(config_file, checkpoint_file, device='cuda:0')

# Test a single image 
img = "/content/mmdetection/CascadeTabNet/Demo/demo.png"

# Run Inference
result = inference_detector(model, img)

# Visualization results
show_result_pyplot(img, result,('Bordered', 'cell', 'Borderless'), score_thr=0.85)
```

```
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
import mmcv

# Load model
config_file = '/content/CascadeTabNet/Config/cascade_mask_rcnn_hrnetv2p_w32_20e.py'
checkpoint_file = '/content/epoch_36.pth'
model = init_detector(config_file, checkpoint_file, device='cuda:0')

IMG_PATH = r"/content/images/"

## loop over test images
files = os.listdir(IMG_PATH)
cont=0
for file in sorted(files):
  if file.endswith(".jpg") or file.endswith(".png") or file.endswith(".tif"):

    # Test a single image 
    img = IMG_PATH + file

    # Run Inference
    result = inference_detector(model, img)

    # Visualization results
    show_result_pyplot(img, result,('Bordered', 'cell', 'Borderless'), score_thr=0.85)
```
